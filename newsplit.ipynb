```python
import json
import numpy as np
import pandas as pd
import ast
```

```python
business_data = []
with open('business.json') as f:
    for line in f:
        business_data.append(json.loads(line))
business_df = pd.DataFrame.from_dict(business_data)
business_df.head()

target = pd.read_csv("restaurantfinal.csv")
```

```python
target
```

```python
business_df = business_df[business_df.business_id.isin(target["business_id"].values)]
```

```python
business_df.reset_index(inplace=True)
business_df.drop("index",axis=1,inplace=True)
```

```python
business_df = business_df.sort_values(by=["business_id"])
```

```python
newdf = pd.DataFrame(business_df['attributes'].to_dict()).T
```

```python
newdf.BusinessParking.replace({np.nan: "None"}, inplace=True)
```

```python
newdf.Ambience.replace({np.nan: "None"}, inplace=True)
```

```python
newdf.BestNights.replace({np.nan: "None"}, inplace=True)
```

```python
newdf.GoodForMeal.replace({np.nan: "None"}, inplace=True)
```

```python
newdf.Music.replace({np.nan: "None"}, inplace=True)
```

```python
newdf['BusinessParking'] = newdf['BusinessParking'].apply(ast.literal_eval)
newdf['Ambience'] = newdf['Ambience'].apply(ast.literal_eval)
newdf['BestNights'] = newdf['BestNights'].apply(ast.literal_eval)
newdf['GoodForMeal'] = newdf['GoodForMeal'].apply(ast.literal_eval)
newdf['Music'] = newdf['Music'].apply(ast.literal_eval)

```

```python
newdf.to_csv("newdf.csv")
```

```python
cols_to_split = ['BusinessParking', 'Ambience', 'BestNights', 'GoodForMeal', 'Music']
for col_to_split in cols_to_split:
    new_df1 = pd.DataFrame(newdf[col_to_split].to_dict()).T
    new_df1.columns = [col_to_split + '_' + str(col) for col in new_df1.columns]
    business_df = business_df.join(new_df1)
```

```python
business_df.to_csv("attributes.csv")
```

```python
# columns with non-boolean categorical values:
cols_to_split = ['AgesAllowed', 'Alcohol', 'BYOBCorkage', 'NoiseLevel', 'RestaurantsAttire', 'Smoking', 'WiFi']
new_cat = pd.concat([pd.get_dummies(business_df[col], prefix=col, prefix_sep='_') for col in cols_to_split], axis=1)
# keep all columns (not n-1) because 0's for all of them indicates that the data was missing (useful info)
business_df = pd.concat([business_df, new_cat], axis=1)
business_df.drop(cols_to_split, inplace=True, axis=1)
business_df.head()
```
