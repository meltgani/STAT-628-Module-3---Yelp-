```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
review = pd.read_csv("review.csv")
review
```

```python
review=review.sort_values(by=['business_id'])
review
```

```python
rest = pd.read_csv("restaurant6.csv")
rest
```

```python
standard=pd.merge(review,rest,on ="business_id")
standard['business_id'].nunique()
standard = standard[['business_id', 'stars_y','Type']]
```

```python
#check skeweness before standardizing
standard['stars_y'].skew()
```

```python
standard.groupby('Type').hist(column='stars_y',ec='black')
```

```python
review.groupby('business_id').agg('var')
```

```python
review=review.groupby('business_id').agg('std')
review
```

```python
standard2=pd.merge(review,rest,on ="business_id")
standard2 = standard2[['business_id', 'stars_x','Type','review_count','stars_y']]
standard2["std_dev"]=standard2["stars_x"]
standard2=standard2.drop(columns=["stars_x"])
standard2
```

```python
review2 = pd.read_csv("review.csv")
review2
```

```python
review_mean=review2.groupby(['business_id'],as_index=False)['stars'].mean()
review_mean
```

```python
standard_final=pd.merge(review_mean,standard2,on ="business_id")
standard_final["avg_stars"]=standard_final["stars"]
standard_final=standard_final.drop(columns=["stars"])
standard_final
```

```python
#calculate the z-score use standard_final
standard_final["z-score"]= (standard_final['stars_y'] - standard_final['avg_stars'])/ (standard_final['std_dev'])
standard_final
```

```python
standard_final.groupby('Type').hist(column='z-score',ec='black')
```

```python
#find the margin of error which is just std/(sqrt of reviews)
standard_final['review_count']=standard_final['review_count'] .astype(float)
standard_final['MOE']=((standard_final['std_dev']) / (standard_final['review_count'])**0.5)

#check the z-score if it is negative then it falls below the mean so subtract the margin of error from stars
# if it is positive then it falls above the mean so we add the margin of error to stars
for i in range(449):
    if standard_final['z-score'][i] > 0:
        standard_final['MOE'][i] = (standard_final['MOE'][i]) * 1
    else:
        standard_final['MOE'][i] = (standard_final['MOE'][i]) * -1

standard_final
```

```python
#create a new column called std_stars
standard_final["std_stars"]= standard_final["stars_y"] + standard_final["MOE"]
standard_final
```

```python
#some ratings were above 5 by a small percentage so cap it
for i in range(449):
    if standard_final['std_stars'][i] > 5:
        standard_final['std_stars'][i]= 5
standard_final
```

```python
standard_final.groupby('Type').hist(column='std_stars',ec='black')
```

```python
#check skewness after standardizing
standard_final['std_stars'].skew()
```

```python
standard_final = standard_final[['business_id', 'std_stars']]
standard_final.to_csv('std_stars')
```
