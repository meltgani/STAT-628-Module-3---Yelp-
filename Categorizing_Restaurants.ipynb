```python
import pandas
```

```python
rest=pandas.read_csv('Restaurants.csv')
```

```python
rest.loc[rest['categories'].str.contains('Chinese'),"Type"] = "Chinese"
rest.loc[rest['categories'].str.contains('Thai'),"Type"]= "Thai"
rest.loc[rest['categories'].str.contains('Japanese'),"Type"] = "Japanese"
rest.loc[rest['categories'].str.contains('Indian'),"Type"] = "Indian"
rest.loc[rest['categories'].str.contains('Pizza'),"Type"] = "Pizza"
rest.loc[rest['categories'].str.contains('Bakery'),"Type"] = "Bakery"
rest.loc[rest['categories'].str.contains('Bakeries'),"Type"] = "Bakery"
rest.loc[rest['categories'].str.contains('Mexican'),"Type"] = "Mexican"
rest.loc[rest['categories'].str.contains('Burgers'),"Type"] = "Burgers"
rest.loc[rest['categories'].str.contains('Sushi'),"Type"] = "Japanese"
rest.loc[rest['categories'].str.contains('Coffee'),"Type"] = "Coffee"
rest.loc[rest['categories'].str.contains('Ice Cream'),"Type"] = "Ice Cream & Frozen Yogurt"
rest.loc[rest['categories'].str.contains('Mediterranean'),"Type"] = "Mediterranean"

rest
```

```python
bus=rest.loc[rest["is_open"] ==1] #remove closed business
bus2=bus[bus['categories'].str.contains('Hotels')==False].reset_index() #drop hotels
rest1=bus2[bus2.address.notnull()] #drop restaurants with no address
rest1

```

```python
rest1.to_csv('restaurant1.csv')
```
