```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
```

```python
stop_words = list(stopwords.words('english'))
```

```python
import pandas as pd
```

```python
business_df = pd.read_csv("Final_Business_0_1.csv")
```

```python
business_df.drop("Unnamed: 0",axis = 1, inplace =True)
```

```python
review = pd.read_csv("review.csv")
```

```python
review1 = review[review.business_id.isin(business_df.business_id.values)]
```

```python
len(review1.business_id.unique())
```

```python
business_df.Type.unique()
```

```python
Italian = business_df[business_df.Type=="Italian"]
```

```python
Italian_review = review1[review1.business_id.isin(Italian.business_id.values)]
```

```python
Italian_review
```

```python
Italian_review["text"] = Italian_review["text"].str.lower()
```

```python
def cleaning_stopwords(text):
    return " ".join([word for word in str(text).split() if word not in stop_words])
Italian_review['text'] = Italian_review['text'].apply(lambda text: cleaning_stopwords(text))
```

```python
Italian_review.iloc[0].text
```

```python
import string
import re
english_punctuations = string.punctuation
punctuations_list = english_punctuations
def cleaning_punctuations(text):
    translator = str.maketrans('', '', punctuations_list)
    return text.translate(translator)
Italian_review['text']= Italian_review['text'].apply(lambda x: cleaning_punctuations(x))

def cleaning_repeating_char(text):
    return re.sub(r'(.)1+', r'1', text)
Italian_review['text'] = Italian_review['text'].apply(lambda x: cleaning_repeating_char(x))

def cleaning_URLs(data):
    return re.sub('((www.[^s]+)|(https?://[^s]+))',' ',data)
Italian_review['text'] = Italian_review['text'].apply(lambda x: cleaning_URLs(x))

def cleaning_numbers(data):
    return re.sub('[0-9]+', '', data)
Italian_review['text'] = Italian_review['text'].apply(lambda x: cleaning_numbers(x))
```

```python
Italian_review['text'].iloc[0]
```

```python
from nltk.probability import FreqDist
from nltk.tokenize import word_tokenize
```

```python
Italian_review_text = Italian_review.text.str.cat(sep = " ")
tokens = word_tokenize(Italian_review_text)
fdist = FreqDist(tokens)
fdist.most_common()
bigrams_list = ["_".join(item) for item in nltk.bigrams(tokens)]
trigrams_list = ["_".join(item) for item in nltk.trigrams(tokens)]
FreqDist(bigrams_list).most_common()
FreqDist(trigrams_list).most_common()
```

```python
bad_Italian=Italian[Italian.stars<3.5]
bad_Italian_review = Italian_review[Italian_review.business_id.isin(bad_Italian.business_id.values)]
bad_Italian_review_text = bad_Italian_review.text.str.cat(sep = " ")
bad_italian_tokens = word_tokenize(bad_Italian_review_text)
FreqDist(bad_italian_tokens).most_common()
bigrams_list_bad_italian = ["_".join(item) for item in nltk.bigrams(bad_italian_tokens)]
trigrams_list_bad_italian = ["_".join(item) for item in nltk.trigrams(bad_italian_tokens)]
# FreqDist(bigrams_list_bad_italian).most_common()
FreqDist(trigrams_list_bad_italian).most_common()
```
