```python
import pandas
```

```python
final = pandas.read_csv("restaurant_almost_final.csv")
final
```

```python
final[['opening','closing']] = final.hours.str.split('-',expand=True)
final['opening'] = final['opening'].str.replace(":",".")
final['closing'] = final['closing'].str.replace(":",".")
final
```

```python
final["opening"]=final.opening.astype(float)
final["closing"]=final.closing.astype(float)

final["hours_per_day"]=abs(final['closing'] - final["opening"])
final
```

```python
total=final.groupby(["business_id"])["hours_per_day"].sum()
total.columns=['business_id','hours_per_day']
total=total.reset_index()
total
```

```python

final_hours=pandas.merge(total,final,on ="business_id")
final_hours["hours_per_week"]=final_hours["hours_per_day_x"]
final_hours=final_hours.drop(columns=["Unnamed: 0","hours","day_of_the_week","opening","closing","hours_per_day_y","hours_per_day_x"])
```

```python
final_hours=final_hours.drop_duplicates()
final_hours
```

```python
final_hours.to_csv('restaurant6.csv')
```
