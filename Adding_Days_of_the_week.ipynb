```python
import pandas as pd ###
import numpy as np
from pathlib import Path 
```

```python
data = pd.read_csv("restaurant4.csv") ###
data.head()
data = data.drop(['name','address','state','latitude','longitude','categories','is_open','attributes'],axis=1)
data.head(8)
#len(data)
data.business_id.unique().size
```

```python
new_df = pd.DataFrame(data.hours.str.split(',').tolist(), index=data.business_id).stack() ###
new_df = new_df.reset_index([0, 'business_id'])
#new_df = ['business_id', 'hours']
new_df.columns = ['business_id', 'hours']
new_df.head()
```

```python
new_df[['day_of_the_week','hours']] = new_df['hours'].str.split(': ',expand=True) ###
new_df['day_of_the_week'] = new_df['day_of_the_week'].str.replace("'"," ")
new_df['day_of_the_week'] = new_df['day_of_the_week'].str.replace("{"," ")
new_df['day_of_the_week'] = new_df['day_of_the_week'].str.replace("}"," ")
new_df['day_of_the_week'] = new_df['day_of_the_week'].str.replace(" ","")
#new_df['day_of_the_week'] = new_df['day_of_the_week'].replace("Monday","m",regex=True)
new_df['hours'] = new_df['hours'].str.replace("'"," ")
new_df['hours'] = new_df['hours'].str.replace("{"," ")
new_df['hours'] = new_df['hours'].str.replace("}"," ")
new_df['hours'] = new_df['hours'].str.replace(" ","")
new_df.head(8)
```

```python
days = ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
new_df['day_of_the_week'] = new_df['day_of_the_week'].astype('string')
count = 0
new_df2 = new_df
for i in new_df['business_id'].unique():
#     temp_days = days
    newdf1 = new_df[new_df["business_id"]==i]
    if len(newdf1)!= 7:
        count += 1
        print(i)
        temp = newdf1.day_of_the_week.str.strip().values
        diff = set(days)-set(temp)
        print(diff)
        for d in diff:
            newrows = {'business_id': i, 'hours': "0:0-0:0", 'day_of_the_week': d}
            new_df2=new_df2.append(newrows,ignore_index=True)
            print(newrows)
```
